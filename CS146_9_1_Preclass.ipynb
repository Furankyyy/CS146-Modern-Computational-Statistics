```
import numpy as np
import scipy.stats as sts
```

```
def generate_samples(n_games,attempts):
    fg = sts.beta(1,1).rvs()
    success = []
    for i in range(n_games):
        success.append(sts.binom.rvs(attempts,fg))
    return fg,success
```

```
n_games = 4
attempts = [5,10,15,20]
```

```
field_goal = []
shots_made = []
for j in range(3):
    fg, success = generate_samples(n_games,attempts)
    field_goal.append(fg)
    shots_made.append(success)
```
