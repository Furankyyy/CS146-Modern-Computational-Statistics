```
import numpy as np
from scipy.stats import gamma, norm, beta
import matplotlib.pyplot as plt


x = np.linspace(0,50,100)
```

```
plt.plot(x,gamma.pdf(x,8))
plt.plot(x,gamma.pdf(x,4),color="red")
```

```
y = gamma.rvs([8],size=100)
plt.hist(y,density=1)
plt.plot(x,gamma.pdf(x,8))
plt.xlim(0,20)
plt.ylim(0,0.4)
```

```
plt.plot(x,beta.pdf(x,5,4))
plt.plot(x,beta.pdf(x,1,8),color="red")
plt.xlim(0,1)
```

```
y = beta.rvs(5,4,size=100)
plt.hist(y,density=1)
plt.plot(x,beta.pdf(x,5,4))
plt.xlim(0,1.2)
```

```
plt.plot(x,norm.pdf(x,30,3))
plt.plot(x,norm.pdf(x,20,20))
```

```
y = norm.rvs(30,3,size=100)
plt.hist(y,density=1)
plt.plot(x,norm.pdf(x,30,3))
```
